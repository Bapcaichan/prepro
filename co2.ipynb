```python
import pandas as pd

df = pd.read_csv(r'C:\Users\ADMIN\Documents\VScode\Python\github\prepro\data\co2.csv')
```

```python
df.info()
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['co2'], label="CO₂ (Trước làm sạch)", color='red', alpha=0.6)
plt.xlabel("Thời gian")
plt.ylabel("Nồng độ CO₂ (ppm)")
plt.title("Biểu đồ CO₂ trước khi làm sạch dữ liệu")
plt.legend()
plt.grid(True)
plt.show()
```
